```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
#Fetching data
Cali_X,Cali_y =datasets.fetch_california_housing(return_X_y=True)
```

```python
Cali_X=Cali_X[:,np.newaxis,2] #slicing the dataset
```

```python
#slicing train and test data
Cali_X_train=Cali_X[:-20]
Cali_X_test=Cali_X[-20:]
Cali_y_train= Cali_y[:-20]
Cali_y_test=Cali_y[-20:]
```

```python
#Linear Regression model
regression=linear_model.LinearRegression()
regression.fit(Cali_X_train,Cali_y_train)
```

```python
Cali_y_pred=regression.predict(Cali_X_test)
```

```python
print("Weight: ",regression.coef_)
```

```python
#checcking on test data
plt.scatter(Cali_X_test,Cali_y_test,color="blue")
plt.plot(Cali_X_test,Cali_y_pred,color="green")
plt.show()
```

```python
#checking on train data
Cali_y_pred=regression.predict(Cali_X_train)
plt.scatter(Cali_X_train,Cali_y_train,color="blue")
plt.plot(Cali_X_train,Cali_y_pred,color="red")
plt.show()
```
