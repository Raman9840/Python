```python
import pandas as pd
import numpy as np
```

```python
nums=[1,2,3,4,5] #creating a list
s=pd.Series(nums) #converting a list into series using pandas
print(s)
```

```python
nums=[1,2,3,4,5]
s=pd.Series(nums,index=[1,2,3,4,5]) #creating custom index
s
```

```python
#creating a pandas series from a dictionary
dct={'name':'Raman','country':'nepal','city':'lalitpur'}
s=pd.Series(dct)
s
```

```python
#creatin a constant pandas series
s=pd.Series(10,index=[1,2,3,4,5])
s
```

```python
#creating a pandas series using linspace
#linspace creats an array evenly spaced values over the defined/specific interval
s=pd.Series(np.linspace(5,20,10)) #linespace(start, end, no. of items)
s
```

```python
#creating dataframes from list of lists
data=[['raman','Neupane','Lalitpur'],
      ['hari','gopal','lele'],
      ['gopal','bhatta','bhaktapur']]
df=pd.DataFrame(data, columns=['Name','Lastname','City'])
df
```

```python
#creating a dataframe using dictionary
data = {'Name': ['Asabeneh', 'David', 'John'], 'Country':[
    'Finland', 'UK', 'Sweden'], 'City': ['Helsiki', 'London', 'Stockholm']}
df = pd.DataFrame(data)
df
```

```python
#creating a dataframe using a list of dictionary
data=[
    {'Name': 'Asabeneh', 'Country': 'Finland', 'City': 'Helsinki'},
    {'Name': 'David', 'Country': 'UK', 'City': 'London'},
    {'Name': 'John', 'Country': 'Sweden', 'City': 'Stockholm'}
    ]
df=pd.DataFrame(data)
df
```
